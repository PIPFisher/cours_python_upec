# L'interface de Jupyter Notebook

On peut lancer Jupyter Notebook, soit à partir du raccourci présent dans le dossier Anaconda3 (XX-bit) du menu Démarrer, soit à partir d'Anaconda Navigator. L'un comme l'autre :

1. Lance le **server local de Jupyter** qui s'affiche sous la forme d'unefenêtre de console (lorsqu'il est lancé à partir du raccourci)



2. Ouvre **l'interface de Jupyter Notebook** dans le navigateur par défaut à l'adresse http://localhost:8888/tree



Elle dispose de 3 onglets de base :

- **Files** : 
    - C'est un onglet qui sert d'explorateur de fichiers. 
    - C'est aussi l'onglet qui permet de créer un nouveau notebook. 
    - Pour ça, il faut aller dans le dossier où l'on veut créer un notebook, puis cliquer sur **New -> Python 3**
- **Running** : onglet qui permet de gérer les kernels actuellement actifs
- **Cluster** : onglet qui permet de gérer des clusters Jupyter

Il peut éventuellement y avoir d'autres onglets qui s'ajoute à ceux-ci si vous installez des modules supplémentaires spécifiques. En particulier, on peut trouver un onglet **Nbextensions** qui affiche la page permettant d'activer/désactiver des extensions.

# Notebooks, kernels, onglets

## Notebooks et onglets

Un notebook est un fichier lisible par Jupyter et contenant à la fois du code et d'autres contenus, vous êtes actuellement en train de lire un notebook.

Un fichier Jupyter Notebook en Python est reconnaissable par son extension **.ipynb** (IPython Notebook)

Même si un fichier **.ipynb** contient du code en Python, il n'est pas directement exécutable par Python contrairement à un fichier **.py**

Un notebook s'ouvre dans un onglet de votre navigateur.

## Kernel et fenêtres

Un **kernel** est le programme qui permet d'exécuter du code informatique. Le seul kernel qui devrait être installé sur votre ordinateur est **Python 3**. 

Dans la suite de ce cours, on parlera, par abus de langage de **kernel** pour parler d'une instance (une session) du kernel Python 3 lancée en arrière-tâche de chacun des notebooks ouverts sur l'ordinateur. 

C'est un kernel qui :
1. exécute le code Python des cellules de code (lorsque l'utilisateur le demande)
2. stocke le contenu des variables utilisées en mémoire.

Etant donné qu'il y a un kernel différent pour chaque notebook ouvert :
- le code exécuté dans un notebook n'impacte pas celui d'un autre notebook
- les variables définies dans un notebook ne peuvent être utilisées depuis d'autres notebooks.
Ils sont **isolés** les uns des autres. 

Le kernel peut être géré depuis le notebook lui-même à l'aide du menu **Kernel**

## Fermeture d'un onglet

<div class="alert alert-block alert-success">
<b>Attention:</b> 
    
- Si vous fermez l'onglet du notebook directement, ça n'éteint pas le kernel (il continue à être actif en mémoire).
- Pour fermer l'onglet (donc le notebook) ET éteindre le kernel, il faut aller sur le menu **File -> Close and Halt**.
- Un autre moyen d'éteindre des kernels est d'aller dans l'onglet **Running** de l'interface principale de contrôle de Jupyter, et d'éteindre les kernels inutiles.
- **N'OUBLIEZ PAS D'ENREGISTRER VOTRE NOTEBOOK AVANT DE LE FERMER**
</div>

# Les cellules

L'intérêt des notebooks est de permettre de mêler des morceaux de codes et des contenus divers et variés (textes, images, titres de sections...).

Le notebook est un peu au code Python ce que le fichier Doc de Microsoft Word est au texte brut...

Mais un notebook n'est au final qu'un ensemble de cellules se suivant à la suite les unes des autres

Ces cellules ont toutes un code sous-jacent indiquant à Jupyter ce qu'il doit en faire

## Mode édition et mode commande

Sur Jupyter Noteboook, 2 modes d'interaction sont possibles :
- Lorsque l'indicateur de couleur à gauche de la cellule sélectionnée est **en bleu**, on est en **mode commande**. On passe alors d'une cellule à une autre avec les flèches haut et bas.

> Si, lorsque l'on est en **mode commande**, on clique sur le bouton **Entrée** du clavier (ou si on double-clique sur la cellule), on passe en **mode édition**

- Lorsque l'indicateur de couleur à gauche de la cellule sélectionnée est **en vert**, on est en **mode édition**. On peut alors modifier le code contenu dans la cellule. 

> Si, lorsque l'on est en **mode édition**, on clique sur le bouton **Echap** du clavier, on revient en **mode commande** (mais sans exécuter le code de la cellule)

## Exécution d'une cellule

Pour exécuter le code de la cellule, on tape (en mode édition ou en mode commande, peu importe) soit :
- Ctrl+Entrée : exécute le code pour une cellule code ou affiche le résultat final d'une cellule de contenus.
- Shift+Entrée : idem, mais passe aussi à la cellule suivante.

Soit en cliquant avec la souris sur le bouton **Run cell** de la barre d'outils (équivalent à Shift+Entrée)



## Types de cellules

Ces cellules peuvent être de trois types :
- Les cellules de **contenu** (texte, titre de section, image, lien web...), appelées **markdown** en anglais
- Les cellules de **code** (de code Python en l'occurrence)
- Les cellules de contenu **brut**, appelées **raw** en anglais, très peu utilisées

### Contenu (Markdown)

Lorsqu'on les exécute, le résultat de leur contenu mis en forme est affiché

Elles peuvent contenir plusieurs sortes de contenu différentes comme nous allons le voir ci-dessous

### Code

Les cellules de code, qui doivent contenir du code en Python écrit en respectant la syntaxe du langage, sont des cellules exécutables.

````python
print('Bienvenue dans ce cours de Python !')
````

Les cellules de code ont un input, et, une fois exécutée, un output, respectivement signifiées par les indications **In \[N\]** et **Out\[N\]** avec **__N__** indiquant qu'il s'agit de la Nième exécution d'une cellule depuis le démarrage du **kernel**

<div class="alert alert-block alert-success">
<b>Astuce:</b> 
    
Dans les cellules de code, lorsque l'on fait une sélection de texte avec le curseur et que l'on clique sur **__(__**, **__[__**, **__{__**, **__'__** ou **__"__**, au lieu de remplacer la sélection par ce caractère, comme on pourrait s'y attendre, Jupyter Notebook transforme cette action en **encadrement de la sélection**, ce qui est très pratique!
</div>



### Contenu brut

Pour information, un troisième type de cellule existe, les cellules de contenu brut (**Raw NBConvert**), elles ne peuvent être exécutées d'aucune manière, elles gardent toujours la même apparence. 

Elles ne sont quasiment jamais utilisées...

## Barre de menus

- **File** : Modification de notebook : Ouvrir, fermer, enregistrer, faire une copie...

- **Insert** : Insérer des cellules ou des titres

- **Kernel** : Redémarrage, interruption du kernel
- **Navigate** : Aller à une section du notebook (sommaire)
- **Widgets** : Utilisation des widgets

Eléments du menu essentiels :

- **File -> Close and Halt** : Pour bien fermer le notebook
- **Edit -> Merge Cell Above/Below** : Pour fusionner la cellule actuelle avec celle juste au-dessus/en-dessous
- **Edit -> Split Cell** : Diviser une cellule en deux cellules là où est le curseur
- **Edit -> nbextension config** : Configuration des extensions
