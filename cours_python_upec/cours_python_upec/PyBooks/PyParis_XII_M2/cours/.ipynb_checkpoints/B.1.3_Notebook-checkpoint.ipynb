# Notebooks, kernels, onglets

## Notebooks et onglets

Un notebook est un fichier lisible par Jupyter et contenant √† la fois du code et d'autres contenus, vous √™tes actuellement en train de lire un notebook.

Un fichier Jupyter Notebook en Python est reconnaissable par son extension **.ipynb** (IPython Notebook)

## Kernel et fen√™tres

Un **kernel** est le programme qui permet d'ex√©cuter du code informatique. Le seul kernel qui devrait √™tre install√© sur votre ordinateur est **Python 3**. 

Dans la suite de ce cours, on parlera, par abus de langage de **kernel** pour parler d'une instance (une session) du kernel Python 3 lanc√©e en arri√®re-t√¢che de chacun des notebooks ouverts sur l'ordinateur. 

C'est un kernel qui :
1. ex√©cute le code Python des cellules de code (lorsque l'utilisateur le demande)
2. stocke le contenu des variables utilis√©es en m√©moire.

Etant donn√© qu'il y a un kernel diff√©rent pour chaque notebook ouvert :
- le code ex√©cut√© dans un notebook n'impacte pas celui d'un autre notebook
- les variables d√©finies dans un notebook ne peuvent √™tre utilis√©es depuis d'autres notebooks.
Ils sont **isol√©s** les uns des autres. 

Le kernel peut √™tre g√©r√© depuis le notebook lui-m√™me √† l'aide du menu **Kernel**
