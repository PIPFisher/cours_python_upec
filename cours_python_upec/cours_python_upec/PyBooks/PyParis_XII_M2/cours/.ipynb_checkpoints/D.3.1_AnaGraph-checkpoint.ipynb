## Courbes de distribution et box plots

##### En nombre d'éléments

On définit le nombre de barres en utilisant le paramètre **bins**

```python
df.price.plot.hist(bins=100)
```

##### En densité

Le graph de distribution précédent présente en ordonnées le nombre d'éléments dans chaque **bin**, mais on peut aussi obtenir la densité en utilisant le paramètre adéquat (density).

```python
df.price.plot.hist(
    bins=100, 
    density=True
)
```

##### Estimation de la courbe de densité

On peut faire une courbe de densité théorique estimée en utilisant plutôt la méthode **density()**, ou son synonyme **kde()** (kernel density estimation)

```python
df.price.plot.density()
```

# Seaborn

**seaborn** est une bibliothèque externe de Python (installée par défaut avec **Anaconda**). Elle donne accès à des modes de représentation graphique supplémentaires très intéressants.

```python
import seaborn as sns
```

## Heatmaps

##### Heatmap appliquée à une matrice de correlations

```python
sns.heatmap(new_df_ts.corr())
```

```python
sns.heatmap(
    100*new_df_ts.corr(), 
    square=True, # spécifie que l'on veut que le graph soit carré
    annot=True, # spécifie que l'on veut que les nombres soient affichés
    fmt='.1f', # spécifie le format en utilisant la codification habituelle
)
```

##### Heatmap appliquée à une table de contingence

```python
sns.heatmap(
    pd.crosstab(df['origin'], df['destination']), 
    annot=True,
    fmt='d',
)
```

## Remarques sur les color maps

On appelle **color map** ou **cmap** l'échelle de couleur utilisée, sur seaborn, la cmap par défaut sur Seaborn est *rocket*. 

On peut inverser (reverse) toute échelle de couleur en ajoutant **\_r** à la fin de son nom.

Matplotlib propose un grand nombre d'échelles de couleur prédéfinies.

Les color maps de Matplotlib sont stockées dans le sous-module **plt.cm**, celles de Seaborn sont dans **sns.cm**

```python
sns.heatmap(
    pd.crosstab(df['origin'], df['destination']), 
    annot=True, 
    fmt='d', 
    cmap=sns.cm.rocket_r
)
```

<div class="alert alert-block alert-info"><b>Pour aller plus loin :</b><br>

## Grilles de graph

##### Découpage par modalité

```python
new_df_ts['dayofweek'] = new_df_ts.index.dayofweek
```

```python
ma_grille = sns.FacetGrid(
    new_df_ts, 
    col='dayofweek', 
    row='season'
)

ma_grille.map(
    plt.plot,
    'count'
)
```

```python
ma_grille = sns.FacetGrid(
    new_df_ts, 
    col='dayofweek', 
    row='season'
)

ma_grille.map(
    sns.regplot,
    'temp', 
    'count'
)
```

##### Découpage par variable

```python
sns.pairplot(
    data=new_df_ts,
    x_vars=['dayofweek', 'season'],
    y_vars=['count', 'temp']
)
```
