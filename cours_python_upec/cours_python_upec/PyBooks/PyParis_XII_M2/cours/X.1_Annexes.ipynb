# Installer un module en utilisant la ligne de commande

Pour installer un module, il faut d'abord ouvrir une console de commande

## Ouvrir une console de commande

Lors de l'installation d'Anaconda, il devrait avoir créé un raccourci nommé **Anaconda Prompt** dans le dossier **Anaconda3 (XX-bit)** du **Menu Démarrer**

Sur Mac, un raccourci similaire devrait avoir été créé

## Installation de module

Sous Anaconda, deux installateurs de bibliothèques coexistent :

- **pip** (appelé PyPi) est l'installateur par défaut d'extension de Python, il est plus complet

- **conda** est l'installateur d'Anaconda, il est plus propre

**Lorsque c'est possible il faut de préférance utiliser conda !**

### Installation d'un module avec conda

Pour **installer un module avec conda**, il faut taper : 
> conda install *nom_du_module*

(en remplacant *nom_du_module* par le nom de bibliothèque que vous souhaitez installer)


### Si le module n'existe pas sur les channel de base de conda

**conda** a un nombre limité de modules disponibles, si lors d'une tentative d'installation, **conda** répond :

"PackageNotFoundError: Packages missing in current channels"

cela veut dire qu'il n'est pas sur les "channel" normaux de conda, il faut alors essayer de trouver un "channel" (un site de source) adapté.

Il faut alors trouver un channel où le module est disponible. Le mieux est de chercher sur Internet. Par exemple si nous cherchions le module ***jupyter_contrib_nbextensions*** ainsi :

Pour installer un module depuis un "channel" particulier

> conda install -c *nom_de_channel* *nom_du_module*

(en remplacant *nom_du_module* par le nom de bibliothèque que vous souhaitez installer)

(en remplacant *nom_de_channel* par un autre channel si nécessaire)
