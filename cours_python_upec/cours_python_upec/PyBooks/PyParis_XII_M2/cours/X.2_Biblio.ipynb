# Bibliographie

## Bases de Python

- *Introducing Python - Modern Computing in Simple Packages*, B.Lubanovic, 2014
    - Le meilleur choix selon moi, ni trop léger (comme le sont les Driscoll), ni trop lourd (comme l'est le Lutz)
    - Contient des exercices
    - Hélas, ne traite pas de Jupyter Notebook

<img src="https://covers.oreillystatic.com/images/0636920028659/lrg.jpg" align="left" style="height:300px"/>

---

- *Jupyter Notebook 101* et *Python 101*, M.Driscoll, 2018
    - De bons livres pour apprendre les bases, un peu courts, ils vont moins au fond des sujets que les suivants
    - Achetables par paiement libre (y compris 0 €) : https://leanpub.com/jupyternotebook101 et https://leanpub.com/python_101
    - Dispos en formats ebook ou PDF

<img src="https://s3.amazonaws.com/titlepages.leanpub.com/jupyternotebook101/hero?1530469532" align="left" style="height:300px"/>
<img src="https://s3.amazonaws.com/titlepages.leanpub.com/python_101/hero?1518074081" align="left" style="height:300px"/>

---

- *Learning Python*, M. Lutz, 5th edition, 2013    
    - Un peu vieillissant et perd du temps à parler de Python 2. 
    - Grosse référence. Très (voire trop) épais.
    - Existe en français sous le nom *Introduction à Python*. Ne l'achetez pas, il n'a pas été réédité depuis 15 ans et est bien trop vieux (il ne parle même pas de Python 3)

<img src="https://covers.oreillystatic.com/images/0636920028154/cat.gif" align="left" style="height:300px"/>

---

- *Python 3 - Apprendre à programmer en Python avec PyZo et Jupyter Notebook*, B.Cordeau, L.Pointal, 2017
    - Je ne l'ai jamais lu, et donc ne peut en garantir le contenu, mais j'en ai plutôt entendu du bien
    - Il a l'avantage d'être en français pour ceux qui y tiennent vraiment
    - Il a l'autre avantage d'être très récent et d'utiliser Jupyter Notebook

<img src="https://www.dunod.com/sites/default/files/styles/principal_desktop/public/thumbnails/image/9782100766369-001-X.jpeg" align="left" style="height:300px"/>

---

- *Python Programming for the Absolute Beginner*, M.Dawson, 3rd edition, 2010
    - Un peu vieillissant. 
    - Très ludique, codage de petits jeux pour apprendre les fonctionnalités de base.

<img src="http://t3.gstatic.com/images?q=tbn:ANd9GcRiTjG1TCnki2KP9wQN4ZaLXiB_QhpjLkpoAEt-1rWoGg70HFtQ" align="left" style="height:300px"/>

---

- *Python Pocket Reference - Python in Your Pocket*, M.Lutz, 5th edition, 2014
    - Pas réellement un livre pour apprendre, plutôt un énorme référentiel.

<img src="https://covers.oreillystatic.com/images/0636920028338/cat.gif" align="left" style="height:300px"/>

## Traitement et analyse de données

- *Python Data Science Handbook*, J.VanderPlas, 2017
    - Un livre très solide qui fait une présentation approfondie des outils principaux sous Python, plus technique que le McKinney.
    - Aborde aussi le machine learning
    - Disponible en libre accès à l'adresse suivante : https://jakevdp.github.io/PythonDataScienceHandbook/

<img src="https://covers.oreillystatic.com/images/0636920034919/lrg.jpg" align="left" style="height:300px"/>

---

- *Python for Data Analysis*, W.McKinney, 2nd edition, 2018
    - Un très bon livre qui aborde le sujet de l'analyse des données sous un angle très pratique. 
    - Livre très récent et donc très à la page.
    - Commence par une présentation de Jupyter.
    - Existe en version française, sous le nom *Analyse de données en Python* mais uniquement dans sa première version (2015).

<img src="https://covers.oreillystatic.com/images/0636920050896/lrg.jpg" align="left" style="height:300px"/>
<img src="https://static.eyrolles.com/img/2/2/1/2/1/4/1/0/9782212141092_h430.jpg" align="left" style="height:300px"/>

---

- *Data Science from Scratch - First Principles with Python*, J.Grus, 2015
    - Un livre qui fait un panorama de l'ensemble des Data Science : un peu d'analyse, un peu de machine learning, un peu de big data.
    - Explications manquant d'approfondissement.
    - Peu orienté sur les outils, bien plus sur les fondations des Data Science.

<img src="https://covers.oreillystatic.com/images/0636920033400/cat.gif" align="left" style="height:300px"/>

## Python (très) avancé

- *Fluent Python*, L.Ramalho, 2015
    - Un très bon livre qui reprend l'ensemble des thèmes évoqués dans le cadre du cours de bases de Python, mais en s'attachant systématiquement à présenter le moyen de coder de manière optimale sur chaque sujet de base du langage.
    - Livre très technique

<img src="https://covers.oreillystatic.com/images/0636920032519/lrg.jpg" align="left" style="height:300px"/>

---

- *The Hitchiker's Guide to Python*, K.Reitz & T.Schlusser, 2016

    - Un très bon livre qui s'applique à présenter comment coder au mieux en Python. La cerise sur le gâteau de l'apprentissage du langage.

<img src="https://covers.oreillystatic.com/images/0636920042921/lrg.jpg" align="left" style="height:300px"/>

---

- *Python Cookbook*, D.Beazley & B.K.Jones, 2013
    - Comme son nom l'indique, livre de recettes, de solutions pour répondre à des problèmes spécifiques.
    - Livre un peu vieillissant. 

<img src="https://covers.oreillystatic.com/images/0636920027072/lrg.jpg" align="left" style="height:300px"/>

---

- *Programming Python*, M. Lutz, 4th edition, 2011
    - Suite de *Introducing Python*, il a les mêmes avantages et inconvéniants que lui

<img src="https://covers.oreillystatic.com/images/9780596158118/lrg.jpg" align="left" style="height:300px"/>

## Machine Learning

 - *Introduction to Machine Learning with Python - A guide for Data Scientists*, A.C.Müller & S.Guido, 2016
    - Un très bon livre d'introduction au machine learning.
    - Existe en version française sous le nom *Le machine learning avec Python - La bible des data scientists*

<img src="https://covers.oreillystatic.com/images/0636920030515/cat.gif" align="left" style="height:300px"/>
<img src="https://static.eyrolles.com/img/2/4/1/2/0/3/4/4/9782412034460_h430.jpg" align="left" style="height:300px"/>

<img src="https://covers.oreillystatic.com/images/0636920052289/lrg.jpg" align="left" style="height:300px"/>

    - Un autre très bon livre, assez technique sur le machine learning.
