# Principe général

On peut nommer formats itératifs les formats organisés ligne à ligne de type tables de données. 

Il existe un grand nombre de formats de stockage de bases de données.

On peut mettre en avant plusieurs caractéristiques essentielles des différents formats :
- **Lisible** : peut-il être lu sans utiliser de programme spécifique ? (en utilisant Notepad++ par exemple)
- **Econome** : est-il plus ou moins optimisé en termes de taille sur le disque ?
- **Types** : Conserve-t-il bien les différents types de données utilisables ? En particulier les catégories et dates...
- **Requêtes** : peut-on extraire une partie des données sans charger au préalable l'ensemble de la base de données en mémoire ?
- **Multi-table** : peut-on mettre plusieurs tables dans un seul fichier ?
- **Métadonnées** : peut-on facilement adjoindre des métadonnées ?

# La famille des CSV/TSV, format le plus simple

**CSV : Comma Separated Values**

Le format CSV est caractérisé par :
- Une ligne pour chaque point de donnée

- Un délimiteur pour séparer les colonnes : la virgule (*comma*) dans le monde anglo-saxo ou le point-virgule (*semi-colon*) dans le monde francophone


**TSV : Tab Separated Values**

Identique aux CSV, à la seule différence que le délimiteur pour séparer les colonnes est le caractère tabulation (\t)

**Caractéristiques :**
- **Lisible** : Oui, très facilement, sur Notepad++, Excel, ou n'importe quoi d'autre (sauf si la base est trop grosse : pas plus d'1 million de lignes sous Excel, pas plus de quelques Go de texte sous Notepad++, selon la mémoire disponible sur l'ordi)
- **Econome** : Non, malgré sa simplicité, il est très gourmand en taille
- **Types** : Non
- **Requêtes** : Non
- **Multi-table** : Non
- **Métadonnées** : Il faut ajouter un fichier d'infos, il est même possible d'avoir des C/TSV sans ligne de titres (ou d'en avoir à l'inverse plusieurs, ce qui pose d'autres problèmes)

# Les tables dans les formats Microsoft Excel

Les données Excel sont facilement exploitables, d'autant plus si vos données sont localisées proprement dans un onglet (première ligne, première colonne en haut à gauche, sous forme de tableau...). **pandas**, le module que nous étudierons par la suite permet de facilement les lire et les écrire.

A noter que les fichiers XLS de génération postérieure à 2003 (**XLSX, XLSM**, en opposition à la génération 97-2003 : **XLS**) sont intrinsèquement des fichiers au **format ZIP** (avec une extension trompeuse) contenant des **fichiers XML** lisibles à l'oeil nu (ceci est vrai aussi pour les fichiers **DOCX, DOCM, PPTX, PPTM**).

Autrement dit, en utilisant le cours sur les formats ML, vous devriez déjà être capables d'exploiter à la main des fichiers Microsoft (même si, attention à **ne pas réinventer la roue**, il existe aussi des modules pour faire ça, comme **xlrd**, **openpyxl** ou **xlwings** pour Excel). Pour ceux qui voudraient essayer, il faut commencer par ouvrir le fichier Microsoft en utilisant la libraire **zipfile**.

A noter qu'il est beaucoup plus compliqué d'accéder à un fichier **Microsoft Access**. Il faut utiliser le module **pyodbc**  installé avec Anaconda...

**Caractéristiques :**
- **Lisible** : Oui, facilement sur Excel
- **Econome** : Assez moyen, le seul vrai avantage c'est qu'il est intrinsequement zippé
- **Types** : Une partie des types spéciaux (pas les catégories, mais les dates par exemple)
- **Requêtes** : Non
- **Multi-table** : Oui
- **Métadonnées** : Oui, il suffit d'ajouter un onglet

Un problème est par contre que chaque onglet est limité à 1 millions de lignes

# Le format SQL

Le **SQL** (Structured Query Language) est un langage relativement simple syntaxiquement permettant d'envoyer une requête à une base de données pour obtenir des données structurées en retour. En anglais il se prononce Si-cou-eul comme le mot sequel (signifiant une suite, une continuation).

Le **format SQL** est l'un des piliers du traitement de base de données actuellement, mais cet acronyme recouvre un ensemble de systèmes hérités très différents les uns des autres :
- Il y a les systèmes SQL reposant sur de la base de données **embarquée** (autrement dit stockée sur la machine du programme qui l'utilise, en local), c'est le cas de SQLite, le système que nous utiliserons dans ce cours.
- Il y a les systèmes SQL reposant sur de la base de données **centralisée** avec droits d'accès (plusieurs ordinateurs pouvant accéder à une base de donnée stockée sur un server), c'est le cas de Oracle, MySQL, PostgreSQL ou SQL Server (ces sytèmes ayant des caractéristiques différentes).

Et il y a aussi les systèmes **NoSQL** reposant sur de la base de données **décentralisée** et à structuration plus libre. Cette catégorie a été créée et s'est développée au cours de la dernière décennie en réponse à l'explosion de la quantité de données à traiter (cf. Google Big Query (GBQ) qui est une forme hybride de NoSQL)

Le SQL est un véritable sujet en soi et ne sera que très brièvement abordé ici... De plus nous n'aborderons que le système SQL le plus simple, le traitement de base de données embarquée, gérable en Python en utilisant le module de la bibliothèque standard **sqlite3**.

Mais bien sûr Python permet d'exploiter l'ensemble de ces formats, à travers des bibliothèques dédiées (mais pas installées de base sur **Anaconda**) : **mysql**, **psycopg** (pour PostGreSQL), **pymssql** (pour SQL Server), **redis** (entre autres, pour le NoSQL)...

**Caractéristiques :**
- **Lisible** : Il est nécessaire d'utiliser un système de gestion de bases de données (voire un système de connexion et d'identification)
- **Econome** : Correct
- **Types** : Assez moyen, sauf si on s'y connait bien en SQL
- **Requêtes** : Oui, très bien fait et bien documenté
- **Multi-table** : Oui
- **Métadonnées** : Oui, il suffit d'ajouter une table

# Les autres formats binaires

Les formats **binaires** (dont le SQL fait partie) sont des formats qui ne sont par définition **pas directement lisibles** à l'oeil humain (en utilisant Notepad++ par exemple, vous constaterez qu'un fichier db est constitué de chaines de caractères entourées de caractères headecimaux, autrement dit, de binaire). Ils sont en revanche par construction, bien plus économes en termes d'occupation disque.

## HDF5

HDF est un format performant de stockage de données et c'est l'un des plus compatibles avec pandas. On peut le voir comme une sorte de concurrent de SQL.

**Caractéristiques :**
- **Lisible** : Il est nécessaire d'utiliser un système de gestion de bases de données.
- **Econome** : Correct
- **Types** : Excellent et parfaitement compatible avec Pandas
- **Requêtes** : Oui, très bien fait
- **Multi-table** : Oui
- **Métadonnées** : Oui, il suffit d'ajouter une table

## Pickle

C'est le format officiel de Python, il permet d'écrire **n'importe quel objet de Python**. 

Son principal (peut-être seul) inconvénient est qu'il ne peut pas faire l'objet de requêtes.

**Caractéristiques :**
- **Lisible** : Il est nécessaire d'utiliser un système de gestion de bases de données
- **Econome** : Excellent
- **Types** : Excellent et parfaitement compatible avec Pandas
- **Requêtes** : Non
- **Multi-table** : Oui, et même multi-objets
- **Métadonnées** : Oui, il suffit d'ajouter une table
