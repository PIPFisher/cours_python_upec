# Internet

## Bases

Attention, il ne faut pas confondre **Internet** et **le Web**.

**Internet** est le réseau global par lequel transitent les données informatiques dont **l'une des applications est le Web**

A l'origine de la création d'Internet, il y a son ancêtre, ARPANET (Advanced Research Projects Agency Network), développé par une agence militaire américaine, le DARPA (Defense Advanced Research Projects Agency).

Historique rapide :
- En **1969**, le permier message est envoyé (login) entre deux ordinateurs de Stanford et de l'UCLA, utilisant ARPANET
- En **1983**, les concepteurs d'ARPANET adoptent le protocole **TCP/IP**, Internet tel qu'on le connait aujourd'hui est né véritablement à ce moment-là.
- Entre **1990 et 1992**, le World Wide Web est crée.

## Protocoles de communication

Afin de pouvoir communiquer à travers un réseau, il faut utiliser un protocole, c'est-à-dire un ensemble de règles permettant aux différentes machines de se comprendre.

Les protocoles sont un peu au réseau ce que le code de la route est au réseau routier.

### Couches de protocoles

Toute communication entre deux machines à travers un réseau (même local) implique la transmission de **paquets** de données entre ces machines. Ces **paquets** ont un contenu brut, mais aussi d'autres infos, entre autres une **adresse locale** et une **adresse distante**.

Pour aller d'un ordinateur **local** (notre ordinateur) à un ordinateur **distant** (l'ordinateur auquel on est en train de se connecteur), le paquet transite généralement à travers un **réseau d'autres ordinateurs intermédiaires**.

Pour simplifier, pour organiser ce déplacement de paquets, quatre **ensembles de protocoles** ont été établis :
- Les protocoles de la couche **Accès réseau**, qui ne nous intéressent pas, il s'agit de la partie très technique du réseau (wifi, ethernet, bluetooth)
- Les protocoles de la couche **Internet** (IP) et de la couche **Transport** (TCP), sur lesquels on va s'arrêter brièvement
- Les protocoles de la couche **Application**, qui nous intéresseront directement

### Bases concernant TCP/IP

**TCP/IP** (Transfer Control Protocal et Internet Protocol) est le protocole d'accès réseau et de transport utilisé sur Internet. Il organise l'acheminement des paquets.

Une **adresse IP** peut-être considérée comme une adresse postale.

Elle est composée de **plusieurs sous-parties séparées par des points**, allant **du plus global au plus local**, un peu comme si on écrivait une adresse postale sous la forme France.Creteil.Avenue_du_Général_de_Gaulle.61

Il y a deux générations d'adresses IP : 
- **IPv4** qui est constitué d'adresses du type **aaa.bbb.ccc.ddd**, avec aaa, bbb, ccc et ddd étant des **nombres décimaux entre 0 et 255** (256 possibilités, donc un octet, donc 8\*4 = 32 bits de possibilités en tout, donc 4,3 milliards d'adresses possibles). *192.168.12.15* est une adresse IPv4 valide
- **IPv6**, la nouvelle génération, qui est constitué d'adresses du type **AAAA.BBBB.CCCC.DDDD.EEEE.FFFF.GGGG.HHHH** avec AAAA, BBBB, etc. étant des **nombres hexadécimaux entre \x0000 et \xFFFF** (65536 possibilités, donc deux octets, donc 16\*4 = 128 bits de possibilités en tout, donc 3,4\*10^38 milliards d'adresses possibles). *FD12.330E.50C5.1228.AB12.1589.ABCD.EE57* est une adresse IPv6 valide

Tout ordinateur connecté à **Internet** (à travers un fournisseur d'accès à Internet, le provider) se voit assigner une adresse **IP publique** (qui peut être statique, donc fixée pour toujours par le proivider, ou dynamique, donc changeant à chaque nouvelle connexion au provider). Pour savoir quelle est son adresse IP publique, rien de plus facile, plein de site proposent de vous le dire :

http://www.mon-ip.com/

L'adresse qui vous est donnée est l'adresse de l'ordinateur qui se connecte directement au **fournisseur d'accès à internet** (ou **provider**).

Mais attention, en réalité, vous êtes aussi sur un **réseau local** (celui de l'université de votre entreprise ou encore votre réseau domestique) sur lequel vous avez une autre adresse, sur le réseau local, une adresse dite locale du type 192.168... 

NB: il existe aussi une adresse spéciale, **127.0.0.1**, ayant pour alias **localhost**, il s'agit de "l'adresse internet de la machine pour elle-même" (un peu comme si vous aviez un petit nom pour vous envoyer une lettre à vous-même).

Cela fait que sur un réseau universitaire ou professionnel, l'ordinateur distant ne connait pas votre adresse finale sur le réseau local, il ne connait que l'adresse IP de l'ordinateur servant d'ordinateur de connexion à Internet. **Cela ne signifie pas que vous êtes anonyme**, car cet ordinateur de connexion sait potentiellement tout des connexions des ordinateurs de son réseau local à internet (à travers lui-même)

Attention, à l'inverse de ce qui se passe dans le cas des adresses postales, il est presque impossible d'envoyer et de recevoir des paquets sans que les noeuds du réseau ne connaissent vos adresses IP. Le serveur distant connait votre adresse publique, et le point de connexion à Internet de votre réseau local connait votre adresse locale.

On peut rajouter un intermédiaire entre l'ordinateur distant et l'ordinateur local, un VPN, un proxy, le réseau TOR, qui peuvent accroître l'anonymisation de l'ordinateur local. 

Mais attention dans le cas d'un VPN ou d'un proxy, il y a toujours une machine qui au moment de la transmission vous identifie. Dans le cas du réseau TOR, non, il s'agit d'un réseau d'anonymisation particulièrement performant, mais pas évident à mettre en place et utiliser.

## La couche application 

Il y a un deuxième ensemble de protocoles qui existent, les protocoles de la couche application, c'est un peu la partie émergée de l'iceberg

### Protocoles courants

Vous en connaissez au moins un, le **HTTP** (HyperText Transfer Protocol) qui est le protocole utilisé pour accéder au **Web**.

Mais vous en connaissez peut-être d'autre :
- **FTP** (File Transfer Protocol), utilisé pour accéder aux disques durs d'une machine distante
- **SMTP** (Simple Mail Transfer Protocol) et **POP** (Post Office Protocol), utilisés pour obtenir et envoyer directement des mails à partir d'un logiciel de messagerie (Thunderbird, Outlook, etc.)
- **SSH** (Secured SHell), utilisé pour exécuter directement des instructions sur un ordinateur distant à partir d'une console
- **DNS** (Domain Name Server), utilisé pour transformer des noms de domaine (e.g. fr.wikipedia.org) en adresse IP (e.g. 91.198.174.192)

### Port de connexion

Etant donné qu'une machine peut avoir plusieurs fonctions réseau en même temps, on précise un port de connexion au moment où l'on accède à cette machine (par défaut **80 pour le HTTP**, et **443 pour le HTTPS**).

### Requêtes et réponses, servers et clients

Tout échange entre 2 ordinateurs à travers un réseau se fait toujours sous la forme de deux types possibles d'opérations de transfert :
- des **requêtes** qu'un ordinateur envoie à un autre ordinateur à travers le réseau
- des réponses que ces ordinateurs fournissent en retour à ces **requêtes**

La plupart des protocoles sur Internet (à l'exception notable des protocoles de Peer to Peer) sont **asymétriques**, c'est-à-dire qu'il y a de manière pérenne:
- un ou des ordinateurs **server** qui sont là pour fournir des réponses à des requêtes
- un ou des ordinateurs **clients** qui accedent aux servers pour leur envoyer des requêtes

### Le protocole HTTP

Dans le cadre du protocole **HTTP**, la relation entre deux machines est asymétrique: 
- Le **server** est l'ordinateur sur lequel est stocké le site web. 
- Le **client** est l'utilisateur qui le contacte pour utiliser ce site web.

De notre point de vue (en tant que client), l'échange se fait toujours sous la forme de deux types possibles d'**opérations de transfert** :
- de **requêtes** que nous, clients, envoyons au server
- de **réponses** que le server nous renvoie à nous, clients

NB : HTTPS est la version sécurisée de HTTP, il est nécessaire d'utiliser cette version améliorée du protocole pour transférer des données sensibles (données de carte de crédit par exemple)

# Le Web

## Les trois langages du Web

Il y a trois langages principaux utilisé du côté client dans le cadre de l'utilisation d'une page Web.

### HTML

Le HTML permet de coder le **contenu** d'une page web. Il s'agit plus d'une structure de données qu'un langage à proprement parler.

```html 
<!DOCTYPE html>
<html>
    <head>
        <title>Page Title</title>
    </head>
    <body>

        <h1>This is a Heading</h1>
        <p>This is a paragraph.</p>

    </body>
</html> 
```

<div class="alert alert-block alert-info"><b>Pour aller plus loin :</b> <a href=https://www.w3schools.com/html/default.asp> Section <b>HTML</b> du site w3schools </a></div>

### CSS

Le CSS permet de coder la **mise en page** d'une page web. Comme le HTML, il s'agit en réalité plus d'une structure de données qu'un langage à proprement parler.

```css
body {
  background-color: lightblue;
}

h1 {
  color: white;
  text-align: center;
}

p {
  font-family: verdana;
  font-size: 20px;
}
```

### Javascript

Le JavaScript permet de coder le **comportement** d'une page web. 

C'est un véritable langage, au même titre que Python, mais un langage spécialisé dans le web client-side (donc bien plus restreint en termes d'utilisations possibles).

```javascript
var x = myFunction(4, 3);   // Function is called, return value will end up in x

function myFunction(a, b) {
  return a * b;             // Function returns the product of a and b
}

function toCelsius(fahrenheit) {
  return (5/9) * (fahrenheit-32);
}
document.getElementById("demo").innerHTML = toCelsius(77); 
```

Il est à noter que JavaScript de base est un langage très compliqué et très lourd à utiliser en pratique. Il est d'ailleurs très rare aujourd'hui que l'on code directement en JavaScript pur.

### Client-side et server-side

Il faut bien distinguer client-side et server-side.

Le code qui s'exécute **client-side**, c'est-à-dire **chez le client** (l'utilisateur du site Web), donc dans son navigateur internet. 

**Javacript** est de loin le principal langage de codage client-side.

Le code qui s'exécute **server-side**, c'est-à-dire le code qui va s'exécuter **sur le server**, en amont de l'envoi d'une page web (lors de sa génération), et le code qui s'exécute à la suite d'une requête. 

Il existe plusieurs langages server-side concurrents :
- Python (typiquement en utilisant les modules Django ou Flask)
- PHP,
- Ruby,
- Java, 
- C#,
- ou encore Javascript (en utilisant Node.js)

## Requêtes et réponses HTTP

### Requête HTTP

Une **requête** est composée de quatre éléments fondamentaux :
- Une **adresse URL** (Uniform Resource Locator) qui est écrite dans la barre URL en haut de votre navigateur
- Son **type**:
    - généralement **GET** (utilisé pour demander une page web en utilisant quelques paramètres) 
    - ou **POST** (utilisé par exemple lorsque l'on envoie ses données après avoir rempli un formulaire)
- Les **paramètres de la requête** (dans le cas d'une requête POST)
- Des **en-têtes** (**headers**, dont les **cookies** sont un élément important), des informations supplémentaires sur la requête


L'**adresse URL** est composé de plusieurs parties :
- Une **partie principale** composée de :
    - un **nom de domaine** (qui est en réalité l'alias d'une adresse IP) "http://nom_de_domaine.extension"
    - éventuellement un **numéro de port** (80 s'il n'est pas spécifié, dans le cas du HTTP), sous la forme : "http://nom_de_domaine.extension:numero_de_port"
    - éventuellement complété par un couple (**nom_utilisateur**, **mot_de_passe**), sous la forme : "http://nom_utilisateur:mot_de_passe@nom_de_domaine.extension:numero_de_port"
- Un **chemin d'accès** permettant de cibler un élément spécifique de l'arborescence du site web visité, sous forme : /dossier/sous_dossier/page_web.html (la page peut ne pas avoir d'extension, ou en avoir d'autres : asp, php, etc.)
- les **paramètres de la requête** (uniquement dans le cas d'une requête **GET ayant des paramètres**), sous la forme : ***?key1=value1&key2=value2&key3=value3***. Dans le cas d'une requête **POST**, les paramètres de la requête sont transmis sans être exprimées dans l'URL.

Exemple de requête à la suite d'une recherche du terme "Python" sur Google

https://www.google.fr/search?q=python&sourceid=chrome&ie=UTF-8

Les **paramètres de la requête** sont passés de manière invisible (cas d'une requête POST) ou de manière visible dans l'URL (cas d'une requête GET). Ces paramètres peuvent être représentés sous forme d'un dictionnaire simple qui, à une chaine de caractère associe une valeur sous forme de chaine de caractères.

Les **en-têtes** sont toujours passés de manière invisible (au sens où ils ne figurent pas dans l'URL). Ces en-têtes peuvent aussi être représentés sous forme d'un dictionnaire simple qui, à une chaine de caractère associe une valeur sous forme de chaine de caractères.

### Réponse HTTP

Une **réponse** est similairement composée de quatre éléments fondamentaux :
- Une **adresse IP** de réponse composée de l'adresse IP publique et du port de réponse (variable)
- Des **en-têtes** (**headers**), des informations supplémentaires sur le contenu de la réponse
- D'un **corps** ou **contenu**, c'est à dire d'un contenu (HTML le plus souvent, mais potentiellement autre : fichier, contenu JSON, etc.) envoyé par le server au client en retour de sa requête
- D'un **code de statut**, code chiffrée indiquant rapidement si tout s'est bien passé ou pas (200 : succès de la requête, 404 : Page introuvable, 300 : Erreur interne du server...)

### La navigation web est sans mémoire

Il faut bien comprendre que la navigation web a été conçue à l'origine **sans mémoire**. 

Lorsque l'on passe d'une page à une autre sur un site, la page actuelle n'a aucune raison de savoir ce qu'il s'est passé sur la page précédente ou les informations que vous avez fournies à un moment ou un autre. 

Les seules **variables d'entrée** utiliséees dans une page HTML sont les en-têtes (dont les cookies) et les paramètres de la requête ayant entrainé cette réponse sous forme de page HTML.
