# Initialisation

## Modules

```python
import requests
import bs4
import time
```

## Constantes

```python
LIEN_BASE = 'https://www.carrefour.fr/promotions'
CLASSE_LIEN_BASE = 'product-card__image-link'
CLASSE_REPONSES = 'nutritional-fact-name'
DOMAINE = 'https://www.carrefour.fr'
```

```python
sess= requests.Session()
```

```python
res=sess.get(LIEN_BASE)
```

```python
res.content[:100]
```

```python
soup = bs4.BeautifulSoup(res.content, 'html.parser')
```

```python
resultats=soup.find_all(class_=CLASSE_LIEN_BASE)
```

```python
len(list(resultats))
```

```python
liste_liens= [elem['href'] for elem in resultats]
```

### Boucle des requêtes

### Boucle des réponses

```python
lien = liste_liens[0]
```

```python
#On ajoute le domaine avant
DOMAINE + lien
```

```python
res=sess.get(DOMAINE + lien) 
```

```python
soup = bs4.BeautifulSoup(res.content, 'html.parser')
```

```python
soup
```

```python
resultats=soup.find_all(class_=CLASSE_REPONSES)
```

## Fonctions

```python
def extraction_donnees(noeud):
    return (
        noeud.h3.text, 
        noeud.span.text
           )
```

```python
extraction_donnees(resultats[0])
```

```python
def scrap_page(lien):
    res=sess.get(DOMAINE+lien)
    soup=bs4.BeautifulSoup(res.content, 'html.parser')
    resultats = soup.find_all(class_=CLASSE_REPONSES)

liste_elements_nutri = []
for elem in resultats:
    liste_elements_nutri.append(extraction_donnees(elem))
    
return liste_elements_nutri
```

```python
import pandas as pd
```

```python
pd.DataFrame(scrap_page(liste_liens[0]))
```
