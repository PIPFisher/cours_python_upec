# Impact de l'âge sur la PP, le coût moyen et la fréquence par groupe de postes

## Prime Pure

```python
df_pp_posteage = \
adulte_essent.groupby(['age_benef', 'grand_poste'])['PP1'].mean().reset_index(name='PP_mean')

sns.set(style="white", rc={'figure.figsize':(10, 10)})
a= sns.lineplot(x='age_benef', y='PP_mean', hue='grand_poste', data=df_pp_posteage)
a.set_xlabel('Age', fontsize='14')
a.set_ylabel('Prime Pure moyenne', fontsize='14')
plt.title("Essentielle : Prime pure moyenne en fonction de l'âge, par groupe de postes", fontsize='14')
```

###### La même chose mais en discriminant aussi le genre (pour chaque groupe de poste)

```python
print(adulte_essent.grand_poste.unique())
```

```python
def graph_age_genre(poste):
    
        df_temp = adulte_essent[adulte_essent['grand_poste']==poste]

        df_pp_posteage = \
        df_temp.groupby(['age_benef', 'INDGENPER_BENEF'])['PP1'].mean().reset_index(name='PP_mean')
        sns.set(style="white", rc={'figure.figsize':(10, 10)})
        sns.lineplot(x='age_benef', y='PP_mean', hue='INDGENPER_BENEF', data=df_pp_posteage,  
                     palette=["#E91E63", "#2196F3"])
        plt.legend(title="Genre", loc='upper left', labels=["Femme", "Homme"], fontsize='12')
        plt.title(poste + ": Prime pure moyenne par âge et genre", fontsize='20')

        pass
```

```python
graph_age_genre('Soins courants')
```

```python
graph_age_genre('Dentaire')
```

```python
graph_age_genre('Optique')
```
