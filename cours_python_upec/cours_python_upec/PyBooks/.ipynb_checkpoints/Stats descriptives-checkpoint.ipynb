## Répartition des primes pures annuelles par groupe de postes

```python
def ppa_gp(df, title, list_colors):
    temp = pd.DataFrame(df.groupby('grand_poste')['PP1'].sum().reset_index())
    temp = temp.sort_values(by=['PP1'], ascending=False)
    # Les 5 premiers
    temp2 = temp[:5].copy()
    # Les restant
    new_row = pd.DataFrame(data={
                                'grand_poste' : ['Autres'], 
                                'PP1' : [temp['PP1'][5:].sum()]
                                })
    temp3 = pd.concat([temp2, new_row])
    # Graph
    colors = list_colors
    plt.pie(temp3['PP1'], labels=temp3['grand_poste'], autopct='%1.1f%%', colors=colors)
    plt.title(title, fontsize='14')
    plt.show()
return
```

# Prime pure et répartition des bénéficiaires par niveau de garantie 

```python
def fpp_gar(df, color1, title):
    # DF 1     
    df1 = pd.DataFrame(df.groupby(['NIVEAU_FINAL2'])['PP_global'].mean().reset_index())

    # DF 2
    temp = df[['NUMPERASS', 'NUMPERADH', 'SCONT', 'CB', 'REG', 'NIVEAU_FINAL2']]
    temp= temp.drop_duplicates()
    df2 = temp.groupby(['NIVEAU_FINAL2']).size().reset_index(name='Freq')

    fpp_gar = pd.merge(df1, df2, how='inner', on='NIVEAU_FINAL2')
    
    import numpy as np
    import matplotlib.pyplot as plt

    fig, ax1 = plt.subplots()

    color = color1
    ax1.set_xlabel('Niveau de garantie', fontsize='12')
    ax1.set_ylabel('Prime Pure moyenne', color=color, fontsize='12')
    ax1.bar(fpp_gar['NIVEAU_FINAL2'], fpp_gar['PP_global'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    plt.xticks(fpp_gar['NIVEAU_FINAL2'])  

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'black'
    ax2.set_ylabel("Nombre d'assurés", color=color, fontsize='12')  # we already handled the x-label with ax1
    ax2.plot(fpp_gar['NIVEAU_FINAL2'], fpp_gar['Freq'], color=color, marker='o', linewidth=2)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped

    plt.title(title, fontsize='12')

    plt.show()

    return
```

## Prime pure par rapport à l'âge

```python
def ppglb_age(df, couleur, title):
    df_ppglb_age = pd.DataFrame(df.groupby(['age_benef'])['PP_global'].mean().reset_index(name='PP_mean'))
    sns.set(style="white", rc={'figure.figsize':(10,6)})
    print(sns.lineplot(x='age_benef', y='PP_mean', data=df_ppglb_age, color=couleur, ci=None))
    print(df[['PP_global', 'age_benef', 'age_adh', 'Zone', 'NIVEAU_FINAL2']].corr(method='pearson'))
    plt.title(title + ": Prime pure moyenne en fonciton de l'âge", fontsize='14')
    plt.xlabel("Âge")
    plt.ylabel("Prime pure moyenne")
return
```

# Prime pure globale par âge et par type de bénéficiaire

```python
def ppglb_agetb(df, title):
    sns.set(style="white", rc={'figure.figsize':(15,10)})
    with sns.color_palette("Set2", 1):
         sns.lineplot(x='age_benef', y='PP_global', hue="type_pers_benef", data=df, linewidth=2, ci=None)
    plt.title(title + ": Prime pure par âge et par type de bénéficiaire", fontsize='14')
    plt.legend(title="Type d'assuré", loc='upper left', labels=['Femme', 'Homme', 'Enfant'], fontsize='14')
    plt.xlabel('Age', fontsize='14')
    plt.ylabel('Prime pure', fontsize='14')    
return
```

# Prime pure par zone géographique

```python
def pp_zone(df, title):
    # DF 1     
    df_temp = pd.DataFrame(df[['Zone', 'PP_global']].groupby(['Zone'])['PP_global'].mean().reset_index(name='PP_mean'))
    
    # DF 2
    df_temp2 = df[['NUMPERASS', 'NUMPERADH', 'SCONT', 'CB', 'REG', 'Zone']]
    df_temp2= df_temp2.drop_duplicates()
    df_temp3 = df_temp2.groupby(['Zone']).size().reset_index(name='Freq')
        
    # Graph 1
    sns.set(style="white", rc={'figure.figsize':(10, 5)})
    with sns.color_palette("PRGn", 11):
        sns.catplot(x="Zone", y="PP_mean", data=df_temp, kind="bar", legend=True)
        plt.title(title + ": Prime pure moyenne par zone géographique", fontsize='14')
        plt.xlabel('Zone', fontsize='14')
        plt.ylabel('Prime pure moyenne', fontsize='14')
    
    # Graph 2
    with sns.color_palette("PRGn", 11):
        sns.catplot(x="Zone", y="Freq", data=df_temp3, kind="bar", legend=True)
    plt.title(title + ": Répartition des bénéficiaires par zone", fontsize='14')
    plt.xlabel("Zone")
    plt.ylabel("Nombre de bénéficiaires")
return
```

```python
def pp_zone2(df, color1, title):
    # DF 1     
    df1 = pd.DataFrame(df.groupby(['Zone'])['PP_global'].mean().reset_index())

    # DF 2
    temp = df[['NUMPERASS', 'NUMPERADH', 'SCONT', 'CB', 'REG', 'Zone']]
    temp= temp.drop_duplicates()
    df2 = temp.groupby(['Zone']).size().reset_index(name='Freq')

    fpp_gar = pd.merge(df1, df2, how='inner', on='Zone')
    
    import numpy as np
    import matplotlib.pyplot as plt

    fig, ax1 = plt.subplots()
    rc={'figure.figsize':(10, 5)}

    color = color1
    ax1.set_xlabel('Zone', fontsize='14')
    ax1.set_ylabel('Prime Pure moyenne', color=color, fontsize='14')
    ax1.bar(fpp_gar['Zone'], fpp_gar['PP_global'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    plt.xticks(fpp_gar['Zone'])  

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'black'
    ax2.set_ylabel("Nombre d'assurés", color=color, fontsize='14')  # we already handled the x-label with ax1
    ax2.plot(fpp_gar['Zone'], fpp_gar['Freq'], color=color, marker='o', linewidth=2)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped

    plt.title(title, fontsize='14')

    plt.show()

    return
```

```python
def pp_coll(df, title):
    df_mean = pd.DataFrame(df[df['College']!='Retraites'].groupby(['College'])['PP_global'].mean().reset_index())
    df_sum = pd.DataFrame(df[df['College']!='Retraites'].groupby(['College'])['PP_global'].sum().reset_index())
    
    colors = ['MediumTurquoise', 'gold', 'lightskyblue', 'PowderBlue', 'LightSteelBlue']
    
    # Graph 1
    sns.set(style="white", rc={'figure.figsize':(10, 5)})
    # with sns.color_palette("PRGn", 11):
    g= sns.catplot(x="College", y="PP_global", data=df_mean, kind="bar", legend=True, color='grey')
    plt.title(title +" : Prime pure moyenne par College", fontsize='13')
    plt.xlabel("College")
    plt.ylabel("Prime pure moyenne") 
    g.set_xticklabels(rotation=90)
    plt.show()

    # Graph 2    
    plt.pie(df_sum['PP_global'], labels=df_sum['College'], autopct='%1.1f%%', colors=colors)
    plt.title(title +  ' : Prime pure annuelle par college', fontsize='14')
    plt.show()
return
```
