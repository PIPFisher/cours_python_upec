```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

### Import Données

```python
df_presta_final = pd.read_pickle(PICKLE + "df_presta_final.pkl")
```

### Graphs

```python
filtre = df_presta_final[(df_presta_final['NBRACT']  > 1)]
df_freq_filt = filtre.groupby(['CODACT']).size().reset_index(name='Freq')
liste_codacte = df_freq_filt[df_freq_filt['Freq'] >= 50]['CODACT'].tolist()
```

```python
def graph(codact):
    for i in codact:
        df_i = df_presta_final[df_presta_final['CODACT'] == i]
        index = df_i.NBRACT.unique().tolist()
        g = df_i.plot.scatter(
                x='NBRACT', 
                y='MNTFRAREE',
                c='VARBASSS',
                colormap='viridis',
                alpha=1, figsize=(10, 6), 
                label = i)
    return g
```

```python
graph=graph(liste_codacte)
```

```python
graph = graph(['HD7', 'HD4', 'HD2', 'IK', 'FJ', 'CP'])
```

```python
def graph2(codact):
    for i in codact:
        df_i = df_presta_final[df_presta_final['CODACT'] == i]
        g = df_i.plot.scatter(
                x='NBRACT', 
                y='MNTLIGDOS',
                c='MNTRMBSS',
                colormap='cool',
                alpha=1, figsize=(10, 6), 
                label = i)
    return g
```
